```python
interaction_counts = df['food_interactions'].value_counts()
print("Interaction type distribution:\n", interaction_counts)
```

```python
missing_mechanism = (
    df.groupby('name')['food_interactions']
      .apply(lambda x: x.isna().mean() * 100)  # percentage missing
      .sort_values(ascending=False)
)
print("Missing food_interactions % per drug:\n", missing_mechanism)
```
